```python
import pandas as pd
df = pd.read_csv('train.csv')
print(df.head())
```

# Step1- Find missing value

```python
df.isnull().sum()
```

```python
(df.isnull().sum() / len(df))*100
```

```python
# Note:- Here the cabin have 77% missing value 
# drop the value having >50% missing value

df.drop('Cabin',axis=1,inplace=True)
```

```python
# Now fill the missing data havinf <50%  eg. Age
df['Age'] = df['Age'].fillna(df['Age'].median())

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum().sum()
```
